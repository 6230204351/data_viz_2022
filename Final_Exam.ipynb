## ตรวจสอบว่าข้อมูลมีค่า Missing หรือไม่

```python
data.isnull().any() #ดูค่า Missing รายคอลัมน์ของตาราง data
#เนื่องจากมี True เกือบทุกคอลัมน์ แสดงว่ามีค่าที่เป็น Missing 
```

```python
data_all = data[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey','title','age']] # เลือกเฉพาะคอลัมน์ที่ต้องการ
data_all
```

เลือกเฉพาะคอลัมน์ที่ต้องการ

```python
data_hospital= data_hospital[['hcode','name','zone_name','province','amphur','tambon','name5']]   # กำหนดให้ตารางชื่อ data_alldb_df โดยทำการเลือกเก็บข้อมูลในตาราง data_alldb จำนวนทั้งหมด 9  คอลัมน์ 
data_hospital 
```
